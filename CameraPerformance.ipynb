Here we want to compare performance of C++ and Python implementation in order to choose a programming language for ROS2 node implementation. The benchmark is based on comparing camera image capture and transfer since it is the most critical part.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Image capture time comparison using OpenCV interface

cpp_image_capture_ms = np.loadtxt('data/camera_opencv_cap_read_cpp.log')
py_image_capture_ms = np.loadtxt('data/camera_opencv_cap_read_py.log')
capture_compress_ms = np.loadtxt('data/opencv_capture_and_compress.log')


fig, ax = plt.subplots(1, 1)
ax.boxplot([cpp_image_capture_ms, py_image_capture_ms, capture_compress_ms])
ax.set_xticklabels(['C++', 'Python', 'Capture + Compress'], rotation=45, fontsize=12)
ax.set_title('Image capture time comparison')
ax.set_ylabel('Time [ms]')
pass
```
