```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (6, 3)

data_simulated = np.genfromtxt('data/navigation_simulated_epuck_0.25_-0.2.csv',delimiter=',')

data_real = []
for index in range(1, 5, 1):
    data_real.append(np.genfromtxt(f'data/navigation_real_epuck_0.25_-0.2_{index}.csv',delimiter=','))

plt.plot(data_simulated[:,1], data_simulated[:,2], 'r-o', alpha=0.5)
for data in data_real:
    plt.plot(data[:,1], data[:,2], 'b-o', alpha=0.5)

plt.xlabel('X position [meters]')
plt.ylabel('Y position [meters]')

plt.savefig('navigation_comparison.pdf', bbox_inches='tight', pad_inches=0.01)
```
